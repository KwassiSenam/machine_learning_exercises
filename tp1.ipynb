```python
#program that reverse a numpy array
def ReverseNp(a):
    lng = a.shape[0]
    inter = a[lng-1]
    a[lng-1] = a[0]
    a[0] = inter
    for i in range(1,lng-1):
        print(a[lng-1])
        a[i] = a[lng-1-i]
        print(a[i])
    return a;

ReverseNp(np.array([5,9,7]))
```

```python
#program that remplace ood numbers with -1
def Remplace(a):
    n = a.shape[0]
    m = a.shape[1]
    for x in range(n):
        for y in range(m):
            if a[x,y]% 2 == 1:
                a[x,y] = -1
    return a;
Remplace(np.array([[5,4,8,6,9,7], [6,9,3,0,4,2]]))
```
