```python
#Data preprocessing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/airline-safety/airline-safety.csv')
display(df)
```

```python
data = df[['incidents_00_14', 'fatal_accidents_00_14']].to_numpy()
```

```python
def plot_data(data):
    fig,ax = plt.subplots(1,1)
    plt.grid()
    plt.scatter(data[:,0], data[:,1])
    ax.set_xlabel('Total number of incidents, 2000-2014')
    ax.set_ylabel('Total number of fatal accidents, 2000-2014')
    ax.set_title('Incidents and fatal accidents for different airlines')
    plt.show()
    
plot_data(data)
```

```python
#Exercise
#1)
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
std_scaler.fit(data)
print("Mean:" ,std_scaler.mean_)
```

```python
std_scaler.transform(data)
```

```python
#2)
from sklearn.preprocessing import RobustScaler
rob_scaler = RobustScaler()
rob_scaler.fit(data)
print(rob_scaler.transform(data))
```

```python
#3)
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(data)
print(min_max_scaler.transform(data))
```

```python
#4) normalization of the dataset
from sklearn.preprocessing import Normalizer
normalizer = Normalizer().fit(data)
normalized_dataset = normalizer.transform(data)
print(normalized_dataset)
```

```python
#PCA
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd

breastCancer = load_breast_cancer()

X=breastCancer.data
y=breastCancer.target
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca_breastcancer = pca.fit_transform(X)
```

```python
len(y)
```

```python
breastCancerDf = pd.DataFrame(data = pca_breastcancer, columns=['principal component 1', 'principal component 2'])

len(breastCancerDf)
```

```python
breastCancerDf['label'] = breastCancer.target.tolist()
breastCancerDf['label'].replace(0, 'benign', inplace = True)
breastCancerDf['label'].replace(1, 'malignant', inplace=True)

breastCancerDf
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 14)
plt.xlabel('Principal Component - 1', fontsize=20)
plt.ylabel('Principal Component - 2', fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset", fontsize=20)
targets = breastCancer.target_names
colors = ['r', 'g']

for target, color in zip(targets, colors):
    indicesToKeep = breastCancerDf['label'] == target
    plt.scatter(breastCancerDf.loc[indicesToKeep, 'principal component 1']
               , breastCancerDf.loc[indicesToKeep, 'principal component 2'],
               c= color, s=50)
    
plt.legend(targets, prop={'size': 15})
```

```python
#Exercise 1
df = pd.read_csv('./water_potability.csv')
display(df)
```

```python
df.isna().sum().sort_values()
```

```python
y = df['Potability']
X  = df.drop(['Potability'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.ensemble import RandomForestClassifier

#calcule MAD
model = RandomForestClassifier(n_estimators=340)

model.fit(X, y)

importances = model.feature_importances_

final_df = pd.DataFrame({"Features": pd.DataFrame(X).columns, 
                         "Importances":importances})
final_df.set_index('Importances')

final_df = final_df.sort_values('Importances')

display(final_df)
```

```python
#plot the barchart
final_df.plot.bar(color='teal')
```

```python
#K-means clustering
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np

from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=6, cluster_std = 0.50, random_state=0)
plt.scatter(X[:,0], X[:,1], s=50);
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
from yellowbrick.cluster import KElbowVisualizer

kmeans=KMeans()
visualizer = KElbowVisualizer(kmeans, k=(4,12))
visualizer.fit(X) #fit the data to the visualizer
visualizer.show() #Finalize and render the figure
```
